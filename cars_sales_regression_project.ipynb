# car sales prediction

## Importing the relevant libraries

```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import seaborn as sns
sns.set(rc={'figure.figsize':(8,5)})
```

## Loading the raw data

```python
# Load the data from a .csv in the same folder
raw_data = pd.read_csv('cars_sales.csv')
raw_data.head()
```

## Preprocessing

```python
raw_data.describe(include='all')
```

### Determining the variables

```python
data = raw_data.drop(['Model'],axis=1) # the 'Model' feature has more than 300 observations, so dropping the model of the car.

data.describe(include='all')
```

### Checking for Multicollinearity

```python
data_cleaned.columns.values
```

```python
# I will use Variance inflation factor VIF to chick for multicollinearity 
from statsmodels.stats.outliers_influence import variance_inflation_factor

# I will only take the numerical variabkes
variables = data_cleaned[['Mileage','Year','EngineV']]

vif = pd.DataFrame()
 
vif["VIF"] = [variance_inflation_factor(variables.values, i) for i in range(variables.shape[1])]
vif["Features"] = variables.columns
```

```python
vif
```

```python
# Since Year has the highest VIF, I will remove it from the model.
data_no_multicollinearity = data_cleaned.drop(['Year'],axis=1)
```

## Create dummy variables

```python
data_with_dummies = pd.get_dummies(data_no_multicollinearity, drop_first=True)
```

```python
data_with_dummies.head()
```

### Rearrange data

```python
data_with_dummies.columns.values # arrange features and target
```

```python
cols = ['Price','log_price', 'Mileage', 'EngineV', 'Brand_BMW',
       'Brand_Mercedes-Benz', 'Brand_Mitsubishi', 'Brand_Renault',
       'Brand_Toyota', 'Brand_Volkswagen', 'Body_hatch', 'Body_other',
       'Body_sedan', 'Body_vagon', 'Body_van', 'Engine Type_Gas',
       'Engine Type_Other', 'Engine Type_Petrol', 'Registration_yes']
```

```python
data_preprocessed = data_with_dummies[cols]
data_preprocessed.head()
```

## Linear regression model

### Declare the inputs and the targets

```python
# The target(s) (dependent variable) is 'log price'
targets = data_preprocessed['log_price']

# The inputs are everything BUT the dependent variable
inputs = data_preprocessed.drop(['log_price','Price'],axis=1)
```

### Scale the data

```python
from sklearn.preprocessing import StandardScaler

# Create a scaler object
scaler = StandardScaler()
# Fit the inputs 
scaler.fit(inputs)
```

```python
inputs_scaled = scaler.transform(inputs)
```

### Train-Test-Split

```python
from sklearn.model_selection import train_test_split

# Split the variables with an 80-20 split and random state
x_train, x_test, y_train, y_test = train_test_split(inputs_scaled, targets, test_size=0.2, random_state=365)
```

### Create the regression

```python
reg = LinearRegression()
# Fit the regression with the scaled TRAIN inputs and targets
reg.fit(x_train,y_train)
```

```python
# I'll store them in y_hat
y_hat = reg.predict(x_train)
```

```python
plt.scatter(y_train, y_hat) # this is to compare the train and predicted y-hat
plt.xlabel('Targets (y_train)',size=20)
plt.ylabel('Predictions (y_hat)',size=20)
plt.xlim(6,13)
plt.ylim(6,13)
plt.show()
```

```python
# residuals
sns.histplot(y_train - y_hat)

plt.title("Residuals PDF", size=18)

# (y_train - y_hat) are negative values imply 
# that y_hat (predictions) are much higher than y_train (the targets).
```

```python
# R-squared of the model
reg.score(x_train,y_train)
```

### Finding the weights and bias

```python
# intecept 
reg.intercept_
```

```python
# weights (coefficients) of the regression
reg.coef_
```

```python
# Regression summary where we can compare them with one-another
reg_summary = pd.DataFrame(inputs.columns.values, columns=['Features'])
reg_summary['Weights'] = reg.coef_
reg_summary
```
